```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
import time
import shutil,os
```

```python
## pitching in NL

## Code to download all the data sets

## Generating the vectors containing all the year numbers
years = np.linspace(1871,2022,152,dtype=int)

## Looping over the years
for year in years:

    ## Initialising the chrome webdriver
    driver = webdriver.Chrome('/Users/yanshen/Desktop/PhD_UIUC/pythonlearning/chromedriver')

    ## Attaching the yeear number in the url to pull the required url
    url = "https://www.fangraphs.com/leaders.aspx?pos=all&stats=pit&lg=nl&qual=0&type=c,3,13,6,24,19,21,17,18,15,45,59&season="+str(year)+"&month=0&season1="+str(year)+"&ind=0&team=0&rost=0&age=0&filter=&players=0&startdate=&enddate="

    ## Opening the url in a seperate chrome window
    driver.get(url)

    ## Attaching a sleep time to selenium so that the page loads completely
    time.sleep(3)

    ## Find the element on the page with the link text function of Selenium
    element = driver.find_element_by_link_text('Export Data')
    driver.execute_script("arguments[0].click();", element)

    ## Again waiting for 5 seconds so that the download is complete
    time.sleep(3)

    ## Closing the browser window
    driver.quit()

    ## Moving the file from downloads folder to the Project Folder
    shutil.move('/Users/yanshen/Downloads/FanGraphs Leaderboard.csv','/Users/yanshen/Desktop/PhD_UIUC/pythonlearning/pitching'+str(year)+'_f_NL.csv')

    ## Printing the year so that we get to know the point till which the extraction has been completed
    print(year)
```
